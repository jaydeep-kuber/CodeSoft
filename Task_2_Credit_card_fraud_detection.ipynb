This dataset is highly unbalanced because to have only 7506 from 12891690. if do  further process with this data it might we get 90% as safe trans due to 1289169 safe trans. we need to fist balance this dataset

Under-Sampling

build a sample dataset which contain normal  destribution of legal and fraud

```python
# taking random sampl form legal data

legal_sample = legal.sample(n=7506)
```

```python
# concate leaga_sample and fraud in one df called balanced_df

balenced_df = pd.concat([legal_sample, fraud], axis=0)
balenced_df.head(5)
```

Model Training

```python
lr_model = LogisticRegression()
lr_model.fit(X_train , y_train)
```

Evalution

Acuuracy Score

```python
X_train_pred = lr_model.predict(X_train)
X_pred_acc = accuracy_score(X_train_pred, y_train)
```

```python
print("Accuracy Score: " , X_pred_acc)
```

```python
# test prediction

X_test_pred = lr_model.predict(X_test)
X_test_acc = accuracy_score(X_test_pred, y_test)
```
